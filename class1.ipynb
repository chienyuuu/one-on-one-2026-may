```python
# 邏輯運算子 (判斷)：and, or, not, >, <, >=, ==, !=, ...

# and（交集）
True and True -> True
True and False -> False
False and True -> False
False and False -> False

# or（聯集）
True or True -> True
True or False -> True
False or True -> True
False or False -> False

# not（反向）
not True -> False
not False -> True
```

```python
True or False and True -> True
False and False or True -> True
True or False and True and False or True and False or True and True -> True
1 == '1' or 1 >= 1 and False or True -> True
```
