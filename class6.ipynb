```python
'''
1. for 迴圈（迴圈跑一半就夠了）
2. 兩兩比較（第一vs.倒數第一、第二vs.倒數第二、第三vs.倒數第三、...）
3. break
4. if / else


為何只需要跑一半的迴圈：
10 // 2 = 5
0/9
1/8
2/7
3/6
4/5

5 // 2 = 2
0/4
1/3
2
'''

s = input()
n = len(s)

# flag
ok = True # 一開始預設「是」回文

for i in range(n // 2): # i = 0, 1, 2, 3, 4, ...
  # i = 0 -> s[0] != s[n - 1]
  # i = 1 -> s[1] != s[n - 2]
  # i = 2 -> s[2] != s[n - 3]
  if s[i] != s[n - 1 - i]: # 不等於
    ok = False # 「不是」回文
    break
  else: # 等於
    continue

if ok:
  print('yes')
else:
  print('no')
```

```python
names = ['amy', 'bob', 'chris']

'''
印出
chris
bob
amy
'''

n = len(names)

for i in range(n-1, -1, -1):
  print(names[i])
```

```python
# 串列搜尋：找到指定元素的索引值
# 串列名稱.index()
# 若沒出現在串列裡面，「不能」用index搜尋

fruit = ['apple', 'banana', 'orange']

pos1 = fruit.index('apple')
pos2 = fruit.index('banana')
pos3 = fruit.index('orange')
# pos4 = fruit.index('kiwi') # error

print(pos1, pos2, pos3)

# print(pos4) # error
```

```python
# 串列計次：計算指定元素出現的次數
# 串列名稱.count()
# 若沒出現在串列裡面，「可以」用count記次

fruit = ['apple', 'banana', 'orange', 'apple']

num1 = fruit.count('apple')
num2 = fruit.count('banana')
num3 = fruit.count('orange')
num4 = fruit.count('kiwi')

print(num1, num2, num3)

print(num4)
```

```python
# 串列元素修改

list1 = [1, 2, 3, 4, 5, 6, 7]
print(list1)

# 把 2 改成 100
list1[1] = 100
print(list1)

# 把 5 改成 200
list1[4] = 200
print(list1)
```
