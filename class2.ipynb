```python
# 複合指定運算子: +=, -=, *=, /=, ...

x = 10
x = x + 10
print(x)

x += 10
print(x)
```

```python
'''
練習：計算複利本金

本金 * (1 + 利率) ^ 年
利率是 2%
印出 「6 年後存款是 {} 」
'''

money = int(input('本金是多少'))
times = (1 + 0.02) ** 6
money *= times  # money = money * times
print(f'6 年後存款是 {money}')
```

```python
# 判斷式

# 單向判斷式：只有 if

'''
練習：密碼輸入判斷

密碼是 1234
若輸入正確 則輸出歡迎光臨
'''

pwd = int(input('猜密碼'))
if pwd == 1234:
  print('歡迎光臨')
```

```python
# 雙向判斷式：if ... else ...

'''
練習：密碼輸入判斷

密碼是 1234
若輸入正確 則輸出歡迎光臨
若輸入錯誤 則輸出密碼錯誤
'''

pwd = int(input('猜密碼'))
if pwd == 1234:
  print('歡迎光臨')
else:
  print('密碼錯誤')
```

```python
# 多向判斷式：if ... elif ... elif ... (else) ...
# 擇一選擇

'''
練習：判斷 bmi

< 18 過輕
18 <= bmi <= 24 正常
> 24 過胖
'''

bmi = int(input())

if bmi < 18:
  print('過輕')
elif 18 <= bmi <= 24:
  print('正常')
else: # elif bmi > 24
  print('過胖')
```

```python
# 巢狀判斷式：判斷式之內包含判斷式
# 沒有層數限制
# 太多會降低程式可讀性、比較難 debug

'''
a, b, c 是三個不同的正整數
用巢狀判斷式找出最大值

hint: 兩兩比較
'''

a = int(input('a: '))
b = int(input('b: '))
c = int(input('c: '))

if a > b:
  # a, c 是最大值候選人
  if a > c:
    print('a')
  else:
    print('c')
else:
  # b, c 是最大值候選人
  if b > c:
    print('b')
  else:
    print('c')
```
