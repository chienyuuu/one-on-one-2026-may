```python
'''
1. range()
  - 一個參數：產生 0 至 (整數值 - 1) 的串列
  - 兩個參數：產生 (起始值) 至 (終止值 - 1) 的串列
  - 三個參數：產生 (起始值) 開始，每次遞增 (間隔值)，直到 (終止值 - 1) 的串列
2. for：計次的迴圈，用於「已知」次數的迴圈
3. while：條件迴圈，用於「未知」次數的迴圈程式
'''

```

```python
# for 變數 in 數列:
#   程式區塊

for i in range(5):
  print(i)

# while 條件式:
#   程式區塊
n = 5
while n > 0:
  print('hi')
  n -= 1
  # n = n - 1
```

```python
'''
練習：

使用者輸入一個正整數
會輸出1 ~ 該正整數的「總和」
'''
```

```python
'''
練習

使用者輸入一個正整數
用 while 迴圈印出 1 ~ 該正整數的「所有偶數」
'''
num = int(input())

i = 2
while i <= num:
  print(i)
  i += 2
```

```python
# break：強制離開迴圈

# 找最小公倍數
a, b = int(input()), int(input())
max_num = a * b

for i in range(1, max_num+1):
  if (i % a == 0) and (i % b == 0):
    print(i)
    break

# 3, 5 -> 15 (max_num = 15)
# 6, 12 -> 12 (max_num = 72)
```

```python
# continue：迴圈中途暫停不往下執行，直接跳回迴圈起始處繼續執行

# 跳過數字：因為 4 不吉利，所以很多飯店都沒有四樓，可以根據使用者數入的數字，列出有哪些樓層
n = int(input())
for i in range(1, n+1):
  if i == 4:
    continue
  print(i)
```

```python
'''
練習：排除 5 的倍數

使用者輸入一個正整數
印出 1 ~ 該正整數的所有數字，除了 5 的倍數 (除以五的餘數為零)
'''

n = int(input())
for i in range(1, n+1):
  if i % 5 == 0:
    continue
  print(i)
```

```python
# 一維串列宣告
# list1 = [元素1, 元素2, ...]

# data type 可一樣可不一樣
list1 = [1, 2, 3, 4, 5]
list2 = ['apple', 'banana', 'cherry']
list3= [1, 'kiwi', True]

print(list1)
print(list2)
print(list3)

print(list1[1])
print(list2[2])

print(list3[2])
print(list3[-1])
```

```python
# 多維串列
# 串列裡面包串列

list5 = [['amy', 1111], ['bob', 2222], ['chris', 3333]]

print(list5[0])
print(list5[0][0])
print(list5[0][1])

print(list5[2][0])
print(list5[2][1])
print(list5[-1][-1])
```
