```python
'''
1. index()：串列搜尋指定元素的位置，若沒出現在串列裡面，「不能」用index搜尋
2. count()：串列計次指定元素出現的次數，若沒出現在串列裡面，「可以」用count記次
3. 串列修改：在指定位置改成指定元素
'''
```

```python
# zerojudge: https://zerojudge.tw/ShowProblem?problemid=b294

n = int(input())
bread = list(map(int, input().split()))

total = 0

for i in range(5):
  total += bread[i] * (i+1)

print(total)
```

```python
# 串列元素新增

# append()：新增在串列最後面
list1 = [1, 2, 3]
print(list1)

list1.append(100)
print(list1)
```

```python
# insert()：將元素插在指定索引位置，超過就放最後
list1 = [1, 2, 3]
print(list1)

# 在位置 1 插入 100
list1.insert(1, 100)
print(list1)

# 在位置 10 插入 -1
list1.insert(10, -1) # 10 超過現在長度 -> 放最後
print(list1)
```
