```python
'''
1. index()：串列搜尋指定元素的位置，若沒出現在串列裡面，「不能」用index搜尋
2. count()：串列計次指定元素出現的次數，若沒出現在串列裡面，「可以」用count記次
3. 串列修改：在指定位置改成指定元素
'''
```

```python
# zerojudge: https://zerojudge.tw/ShowProblem?problemid=b294

n = int(input())
bread = list(map(int, input().split()))

total = 0

for i in range(5):
  total += bread[i] * (i+1)

print(total)
```

```python
# 串列元素新增

# append()：新增在串列最後面
list1 = [1, 2, 3]
print(list1)

list1.append(100)
print(list1)
```

```python
# insert()：將元素插在指定索引位置，超過就放最後
list1 = [1, 2, 3]
print(list1)

# 在位置 1 插入 100
list1.insert(1, 100)
print(list1)

# 在位置 10 插入 -1
list1.insert(10, -1) # 10 超過現在長度 -> 放最後
print(list1)
```

```python
'''
練習：紀錄一週存款

使用 for 迴圈讓使用者不斷輸入 7 天的存款放進串列
input()中的提示文字要放「請輸入第 ? 天的存款」
再用另一個 for 迴圈累加串列中的金額印出存款的總額
'''

money = []
total = 0

for i in range(7):
  x = input(f'請輸入第 {i+1} 天的存款')
  money.append(int(x))

for i in money:
  total += i

print(total)
```

```python
# 串列元素刪除

# remove()：刪掉串列的指定元素(有重複的會刪第一個)

fruit = ['apple', 'banana', 'cherry', 'apple']

fruit.remove('banana')
print(fruit)

fruit.remove('apple')
print(fruit)

fruit.remove('kiwi') # error!
print(fruit)
```

```python
# pop()：取出並刪除(參數必須是整數)
# 括號內有參數 -> 指定的那個
# 括號內無參數 -> 最後一個

fruit = ['apple', 'banana', 'cherry']

item1 = fruit.pop()
print(item1)
print(fruit)

item2 = fruit.pop(1)
print(item2)
print(fruit)
```

```python
# del：可刪除變數、串列、串列元素

list1 = [1, 2, 3, 4, 5, 6]

del list1[1] # 串列元素
print(list1)

del list1[1:5] # 串列元素(有範圍)
print(list1)

del list1
print(list1) # error!
```

```python
'''
練習：

給一個顏色的串列，請不斷輸入要刪除的顏色，每次都要印出目前剩哪些，當輸入的是 enter 就結束
幫我刪掉橘色、黃色、藍色

hint: while 迴圈不斷輸入、用 break 來結束輸入、用 count 判斷是否存在、用 remove 刪除

color = ['red', 'orange', 'yellow', 'green', 'blue', 'purple']
'''

color = ['red', 'orange', 'yellow', 'green', 'blue', 'purple']

while True:
  x = input('要刪除什麼顏色')

  if x == '':
    break

  if color.count(x) > 0:
    color.remove(x)
  else:
    print('沒有這個顏色')

print('還剩下：', color)
```

```python
# 串列排序

# sort()：由小排到大
list1 = [1, 5, 2, 9, 6, 3]
list1.sort()
print(list1)

list2 = ['k', 'a', 'r', 'b', 'e', 'c']
list2.sort()
print(list2)
```

```python
# reverse()：反轉串列順序

list3 = [1, 2, 3, 5, 6, 9]
list3.reverse()
print(list3)
```

```python
'''
練習：由大排到小

list1 = [1, 5, 2, 9, 6, 3]
'''

list1 = [1, 5, 2, 9, 6, 3]

list1.sort()
list1.reverse()
print(list1)
```

```python
# sorted()
# sort() 是直接改變串列，如果想保留原本的串列就要用 sorted()

list1 = [1, 5, 2, 9, 6, 3]
list2 = sorted(list1)
print(list1, list2)


list3 = sorted(list1, reverse = True)
print(list3)
```

```python
'''
練習：

請不斷成績，每次都要印出目前有哪些，當輸入的是 enter 就結束

hint: while 迴圈不斷輸入、用 break 來結束輸入、用 append 新增進串列、用 sorted 排序
'''

grades = []

while True:
  x = input()

  if x == '':
    break

  grades.append(int(x))

  sort_grades =  sorted(grades)

  print(sort_grades)
```
