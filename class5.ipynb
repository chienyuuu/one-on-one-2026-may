```python
'''
1. break：強制離開迴圈
2. continue：迴圈中途暫停不往下執行，直接跳回迴圈起始處繼續執行
3. 串列：用「中括號」表示
  - 一維串列  list1 = [元素1, 元素2, ...]
  - 空串列  list2 = []
  - 多(二)維串列：串列裡面包串列  list3 = [[], [], []]
'''
```

```python
list3 = [['amy', 1111], ['bob', 2222], ['chris', 3333]]

print(list3[1][1])
print(list3[2][0])


print(list3[2])
print(list3[-1])
```

```python
# 讀取串列元素範圍
# 串列名稱 [起始索引 : 終止索引 : 間隔值]

list1 = [1, 2, 3, 4, 5, 6, 7]


print(list1)
print(list1[2:5]) # 第二項 ~ 第四項
print(list1[0:7:2]) # 奇數
print(list1[1:7:2]) # 偶數
```

```python
# 串列 + for 迴圈

list1 = ['apple', 'banana', 'orange']

# print(list1[0])
# print(list1[1])
# print(list1[2])


for i in list1:
  print(i)
```

```python
# 串列 + for ... range 迴圈

scores = [100, 75, 80]
n = len(scores) # 3

for i in range(n): # 0 ~ n-1
  print(i, scores[i])
```

```python
'''
練習：

names = ['amy', 'bob', 'chris']

印出
編號: 1, 名字: amy
編號: 2, 名字: bob
編號: 3, 名字: chris

'''

names = ['amy', 'bob', 'chris']
n = len(names)

for i in range(n):
  print(f"編號: {i+1}, 名字: {names[i]}")
```

```python
'''
練習：

名字倒印
names = ['amy', 'bob', 'chris']

印出
chris
bob
amy

'''

names = ['amy', 'bob', 'chris']
n = len(names) # 3

for i in range(n-1, -1, -1): # 2, 1, 0
  print(names[i])
```

```python
range(1, 5) -> 1, 2, 3, 4
range(1, 5, 1) -> 1, 2, 3, 4
range(1, 5, 2) -> 1, 3
range(5, 0, -1) -> 5, 4, 3, 2, 1
range(2, -1, -1) -> 2, 1, 0
```
