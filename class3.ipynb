```python
'''
a, b, c 是三個不同的正整數
用巢狀判斷式找出最大值

hint: 兩兩比較
'''

a = int(input('a: '))
b = int(input('b: '))
c = int(input('c: '))

if a > b:
  # 只要比 a, c
  if a > c:
    print('a')
  else:
    print('c')
else:
  # 只要比 b, c
  if b > c:
    print('b')
  else:
    print('c')
```

```python
# range()


# 單一參數

# 數列變數 = range(整數值)
# 產生 0 至 (整數值 - 1) 的串列
list1 = range(10)
print(list(list1))
```

```python
# 兩個參數

# 數列變數 = range(起始值, 終止值)
# 產生 (起始值) 至 (終止值 - 1) 的串列
list2 = range(2, 11)
print(list(list2))

list3 = range(-2, 5)
print(list(list3))

'''
若起始值 >= 終止值，則產生空串列！
'''
list4 = range(5, -2)
print(list(list4))
```

```python
# 三個參數

# 數列變數 = range(起始值, 終止值, 間隔值)
# 產生 (起始值) 開始，每次遞增 (間隔值)，直到 (終止值 - 1) 的串列
list5 = range(2, 12, 1)
print(list(list5))

list6 = range(2, 12, 2)
print(list(list6))

list7 = range(5, -2, -1)
print(list(list7))
```

```python
# for：計次的迴圈，用於已知次數的迴圈

# for 變數 in 數列:
#   程式區塊

for i in range(5):
  print(1)

for i in range(10):
  print(i)
```

```python
'''
練習：

使用者輸入一個正整數
會輸出1 ~ 該正整數的「總和」
'''

x = int(input())

total = 0
for i in range(1, x+1):
  # total = total + i
  total += i

print(total)
```

```python
# 巢狀 for 迴圈

# 會執行很久，會跑一億次 (10000 * 20000)
n = 0
for i in range(1, 10001):
  for j in range(1, 20001):
    n += 1
```

```python
'''
井字直角三角形
輸出格式：

外層 1 次，內層 1 次：#
外層 2 次，內層 2 次：##
外層 3 次，內層 3 次：###
外層 4 次，內層 4 次：####
外層 5 次，內層 5 次：#####
'''

for i in range(1, 6): # [1, 2, 3, 4, 5]
  for j in range(i):
    print('#', end='')
  print()

'''
i = 1
j = 0

i = 2
j = 0, 1

i = 3
j = 0, 1, 2

i = 4
j = 0, 1, 2, 3

i = 5
j = 0, 1, 2, 3, 4
'''
```

```python
# while：條件迴圈，用於未知次數的迴圈程式

# while 條件式:
#   程式區塊

n = 5
while n > 0:
  print(1)
  n -= 1

print(n)

'''
終止條件很重要！！避免寫出無窮迴圈
真的不小心寫出時，可以用 control + c 強制結束程式
'''
```

```python
n = int(input())

total = 0

while n > 0: # 10 -> 9 -> 8 -> ... -> 0
  total += n
  n -= 1

print(total)
print(n)
```

```python
'''
計算階層
1! = 1
2! = 1*2 = 2
3! = 1*2*3 = 6

使用者輸入數字 x，印出 1*2*3*...*x 的結果
'''

x = int(input())

total = 1
while x > 0:
  total *= x
  x -= 1

print(total)
print(x)
```

```python
'''
使用者輸入一個正整數
用 while 迴圈印出 1 ~ 該正整數的「所有偶數」
'''

x = int(input())

i = 0
while i <= x:
  print(i)
  i += 2
```
